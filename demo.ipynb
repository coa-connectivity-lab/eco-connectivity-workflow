# Eco Connectivity Workflow — Demo Notebook

This notebook is a simple demo for the **eco-connectivity-workflow** repo.
It shows basic data loading, a simple spatial check, and placeholder steps for ML and GIS workflows.


```python
import pandas as pd
import numpy as np

print('Environment ready')
```

## 1. Load sample ecological data
We simulate a small dataset representing habitat patches and resistance values.

```python
data = pd.DataFrame({
    'patch_id': [1, 2, 3, 4, 5],
    'x': [10, 20, 30, 40, 50],
    'y': [5, 15, 25, 35, 45],
    'resistance': [0.2, 0.8, 0.5, 0.9, 0.3]
})

data
```

## 2. Simple connectivity logic (placeholder)
We compute a naive connectivity score based on resistance.

```python
data['connectivity_score'] = 1 - data['resistance']
data
```

## 3. ML placeholder step
This represents where resistance models from `ml-models` would be applied.

```python
from sklearn.linear_model import LinearRegression

X = data[['x', 'y']]
y = data['resistance']

model = LinearRegression()
model.fit(X, y)

data['predicted_resistance'] = model.predict(X)
data
```

## 4. GIS export placeholder
In the real workflow, this would be exported to PostGIS / QGIS.

```python
data.to_csv('demo_output.csv', index=False)
print('Exported demo_output.csv')
```
